```python
# Convert images back to video
import ffmpeg

input_images_pattern = '../dataset/test/real/0/%d.png'  # before
output_video = './video/1_resore.mp4'

# Convert images back to video
ffmpeg.input(input_images_pattern, framerate=30).output(output_video, vcodec='libx264', pix_fmt='yuv420p').run()
print('Video created:', output_video)
```

```python
import os
import cv2

# Define input and output directories
input_dir = "./output_images"  # Folder containing the raw output images
output_dir = "./resize_images"  # Folder to save resized images

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Define the target width and height (use sample image dimensions or predefined values)
sample_image_width = 1280  # Replace with desired width
sample_image_height = 704  # Replace with desired height

# Loop through all images in the input directory
for filename in os.listdir(input_dir):
    if filename.endswith(".png") or filename.endswith(".jpg"):  # Modify based on image format
        # Construct the full path for each input image
        input_image_path = os.path.join(input_dir, filename)
        
        # Load the input image
        input_image = cv2.imread(input_image_path)

        # Check if image is successfully loaded
        if input_image is None:
            print(f"Error loading image: {input_image_path}")
            continue

        # Resize the input image to the specified width and height
        input_image_resized = cv2.resize(input_image, (sample_image_width, sample_image_height))

        # Save the resized image to the output directory
        output_image_path = os.path.join(output_dir, filename)
        cv2.imwrite(output_image_path, input_image_resized)

        # Print the dimensions of the resized image
        print(f"Processed {filename}:")
        print(f"Resized image width: {input_image_resized.shape[1]}")
        print(f"Resized image height: {input_image_resized.shape[0]}")

print("Image resizing complete.")
```
